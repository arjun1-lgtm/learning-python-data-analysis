1. The Missing Comma
You write the code my_tuple = ("apple"). What is the data type of my_tuple? How do you fix it so that it is actually a tuple with one element?

```python
my_tuple = ("apple")
print(type(my_tuple))
new_tuple=tuple(my_tuple) # but it is not one element tuple 
print(type(new_tuple))
one_element_tuple=("apple",) #one element tuple , just add a comma 
print(new_tuple)
print(one_element_tuple)
```

2. Direct Access
Given the tuple colors = ("red", "green", "blue", "yellow"), write a line of code to print the last element using negative indexing, and another line to print "green".

3. Simple Unpacking

5. Slicing and Dicing
Given the tuple letters = ('a', 'b', 'c', 'd', 'e', 'f', 'g'):

Write a slice to get ('c', 'd', 'e').

Write a slice to get every second letter, starting from the beginning: ('a', 'c', 'e', 'g').

```python
letters = ('a', 'b', 'c', 'd', 'e', 'f', 'g')
print(letters[2:5])
print(letters[::2])
```

9. Advanced Unpacking (The * Operator)
You have a tuple with 6 elements: scores = (95, 88, 72, 91, 84, 100).
Using Python's extended unpacking (the * operator), unpack this tuple so that the first score goes into highest, the last score goes into latest, and all the middle scores go into a list called others.

```python
scores = (95, 88, 72, 91, 84, 100)
highest,*others,latest=scores
print(highest)
print(latest)
print(others)
```
