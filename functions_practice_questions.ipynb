Question 7: Find the Maximum
Write a function find_max(numbers) that takes a list of numbers and returns the largest one without using the built-in max() function.

```python
def find_max(numbers):
    if not numbers:
        return None
    largest=numbers[0]
    for num in numbers:
        if num>largest:
            largest=num
    return largest 

find_max([])   
find_max([1,2,3,3,34,4,5,])     
```

Question 10 (Counting Characters): Write a function char_count(string) that takes a string and returns a dictionary where the keys are the characters and the values are the number of times each character appears in the string.

```python
def char_count(string):
    # Create an empty dictionary to store frequencies
    counts = {}
    
    # Normalize the string to handle case sensitivity
    clean_string = string.lower()
    
    for char in clean_string:
        # Ignore spaces
        if char == " ":
            continue
        
        # If the character is already in the dictionary, increment its count
        if char in counts:
            counts[char] += 1
        # If the character is new, add it to the dictionary with a count of 1
        else:
            counts[char] = 1
            
    return counts

# Test the function
result = char_count("Hello World")
print(result)
```

Question 3: Prime Checker
Write a function is_prime(n) that returns True if a number is prime, and False otherwise.

Requirements:

Logic: A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

Efficiency: You don't need to check every number up to n. You only need to check up to the square root of n.

Bonus: Once you have that function, try writing another function called get_primes(limit) that returns a list of all prime numbers from 2 up to a given limit.

Example:

is_prime(7) → True

is_prime(10) → False

get_primes(20) → [2, 3, 5, 7, 11, 13, 17, 19]

```python
def is_prime(n):
    # Prime numbers must be greater than 1
    if n <= 1:
        return False
        
    # Check for divisors from 2 up to n-1
    for num in range(2, n):
        if n % num == 0:
            return False  # Found a divisor, so it's not prime
            
    # If the loop finishes without finding a divisor, it is prime
    return True

# Testing it
print(is_prime(7))  # Should print True
print(is_prime(10)) # Should print False
```

Question 4: Palindrome Phrase
Write a function is_palindrome(phrase) that returns True if a string is a palindrome, and False otherwise.

The catch: You must ignore spaces, punctuation (like commas or colons), and capitalization.

Examples:

is_palindrome("racecar") → True

is_palindrome("A man, a plan, a canal: Panama") → True

is_palindrome("hello") → False

```python
def is_palindrome(phrase):
    # 1. Lowercase and remove unwanted characters
    clean_phrase = ""
    for char in phrase.lower():
        if char.isalnum():  # This checks if the character is a letter or number
            clean_phrase += char
            
    # 2. Now compare the clean version to its reverse
    return clean_phrase[::-1] == clean_phrase

# Test it
print(is_palindrome("A man, a plan, a canal: Panama")) # Should be True
print(is_palindrome("racecar"))                       # Should be True
print(is_palindrome("hello"))                         # Should be False
```
