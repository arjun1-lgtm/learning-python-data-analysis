Question 15 .(The Clone Wars): How do you fix the code above so that list_b becomes an independent copy of list_a? That way, when you append 4 to list_b, list_a stays safely as [1, 2, 3].

(Hint: There are a couple of ways—think about slicing or a built-in list method!)

```python
'''hey are not independent at all! This is one of the biggest "gotchas" in Python.

When you write list_b = list_a, you are not copying the data. You are just creating a new shortcut (or reference) to the exact same list in memory.

Here is how you can prove it to yourself. If you check their unique memory IDs using Python's id() function, they match perfectly:'''

list_a = [1, 2, 3]
list_b = list_a

print(id(list_a) == id(list_b))  # Output: True (They are the exact same object!)

'''How to make them truly independent
To make list_b a real, independent clone so that changes to one don't affect the other, 
you have to explicitly tell Python to make a copy. 
Here are the two most popular ways to do it:'''

list_a = [1, 2, 3]
list_b = list_a.copy()  # Makes a brand new copy

list_b.append(4)
print(list_a)  # Output: [1, 2, 3] (Safe!)
print(list_b)  # Output: [1, 2, 3, 4]

#methd 2
list_a = [1, 2, 3]
list_b = list_a[:]  # Slicing from start to finish also creates a copy

list_b.append(4)
print(list_a)  # Output: [1, 2, 3] (Safe!)
```

Question 16: You have a list with a lot of duplicate entries: numbers = [1, 2, 2, 3, 4, 4, 4, 5, 1].

Write a quick one-liner to remove all duplicates from the list, keeping only unique values (the order doesn't matter).

Expected output: [1, 2, 3, 4, 5] (or similar unique collection)


```python
numbers = [1, 2, 2, 3, 4, 4, 4, 5, 1]
values=[]
for num in numbers:
    if num not in values:
        values.append(num)
print(values)  # Output: [1, 2, 3, 4, 5]
```

Question 18: How do you rewrite the grid = ... line using a list comprehension so that it creates 3 independent rows?

Expected Output after changing grid[0][0] = 9: [[9, 0, 0], [0, 0, 0], [0, 0, 0]]

(Hint: You want a list comprehension that creates a [0, 0, 0] row three separate times.)

```python
# grid = [[0] * 3] * 3
# grid[0][0] = 9
# print(grid )

# Create a [0, 0, 0] row, 3 separate times
grid = [[0, 0, 0] for _ in range(3)]

# Now modify just the top-left corner
grid[0][0] = 9

print(grid)
# Output: [[9, 0, 0], [0, 0, 0], [0, 0, 0]]
```

Question 19: You have a list of medals from a race: results = ["Gold", "Silver", "Bronze", "Fourth", "Fifth", "Sixth"].

You want to assign "Gold" to a variable named first, "Silver" to second, and all the remaining runners into a single list variable named the_rest.

Use Python's star unpacking (*) operator to do this in a single line of code.

Expected output when printing the_rest: ['Bronze', 'Fourth', 'Fifth', 'Sixth']

Show me how you use the * operator to capture the rest of the list!

```python
results = ["Gold", "Silver", "Bronze", "Fourth", "Fifth", "Sixth"]

# first gets "Gold", second gets "Silver", the_rest gets everything else
first, second, *the_rest = results

print(first)     # Output: Gold
print(second)    # Output: Silver
print(the_rest)  # Output: ['Bronze', 'Fourth', 'Fifth', 'Sixth']
```

Question 20 (The Matrix Flatten): You have a 2D grid (a list of lists):
matrix = [[1, 2], [3, 4], [5, 6]]
Write a single list comprehension to "flatten" this matrix into one flat list of numbers.

Expected Output: [1, 2, 3, 4, 5, 6]

(Hint: You will need two for loops inside a single set of brackets []. Think about looping through the rows first, and then looping through the items inside each row!)


```python
matrix = [[1, 2], [3, 4], [5, 6]]

# Outer loop: 'for row in matrix' 
# Inner loop: 'for num in row'
flattened = [num for row in matrix for num in row]

print(flattened)
# Output: [1, 2, 3, 4, 5, 6]
```
