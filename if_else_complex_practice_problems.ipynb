#### IF ELSE STATEMENT PRACTICE ####

```python
## SIMPLE CALCULATOR 
NUM1=float(input("Enter first number: "))
NUM2=float(input("Enter second number: "))
operation=input("Enter the  operation (+, -, *, /): ")
if operation=='+':
    result=NUM1+NUM2
elif operation=='-':    
    result=NUM1-NUM2
elif operation=='*':    
    result=NUM1*NUM2
elif operation=='/':    
    if NUM2==0:
        result="Error: Division by zero"
    else:
        result=NUM1/NUM2
else:       
    result="Error: Invalid operation"
print("Result:", result)
```

#### The Autonomous Drone Delivery System (Environmental Checking) ####
An automated delivery drone needs to decide if it is safe to take off and deliver a package.

```python
package_weight=float(input("Enter package weight (in kg): "))
distance=float(input("Enter distance to be shipped (in km): "))
battery_level=float(input("Enter battery level (0-100): "))
weather_condition=input("Enter weather condition (sunny/rainy/snowy): ").strip().lower()
if 0.5<=package_weight<=10 and distance<=100:
    if weather_condition=="stormy" or weather_condition=="snowy":
        take_off=False
    elif weather_condition=="rainy":
        if battery_level>=80 and package_weight<=5:
            take_off=True
        else:
            take_off=False
    elif weather_condition=="sunny":
        if battery_level>=50:
            take_off=True
        elif battery_level>=30 and distance<2:
            take_off=True
        else:
            take_off=False
    else:
        take_off=False                
else :
    take_off=False
if take_off==True:
    print("Drone can take off.")    
else:
    print("Drone cannot take off.")    


```

#### Smart Home Climate & Energy Manager ####
Control a home heating, ventilation, and air conditioning (HVAC) system by balancing comfort with electricity tariff rates


```python
current_temp=float(input("Enter current temperature (in Celsius): "))
target_temp=float(input("Enter target temperature (in Celsius): "))
current_electricity_rate_period=input("Enter current electricity rate (peak, off-peak, shoulder): ").strip().lower()
eco_mode_status=input("Is eco mode enabled? (True/False): ").strip().lower().capitalize()
if current_electricity_rate_period=="peak"and eco_mode_status=="True":
    if abs(current_temp-target_temp)<=3:
        print("Temperature is within the desired range. No action needed,System: AC OFF")
    elif  current_temp>target_temp:
        print("Current temperature is higher than target. System: AC ON")
    else:
        print("Current temperature is lower than target. System: Heater ON")       
else:
    if abs(current_temp-target_temp)<=1:
        print("Temperature is within the desired range. No action needed,System: AC OFF")
    elif  current_temp>target_temp:
        print("Current temperature is higher than target. System: AC ON")
    else:
        print("Current temperature is lower than target. System: Heater ON")    
```

#### Automated Stock Trading Signal Generator ####
Simulate a basic algorithmic trading bot that decides whether to buy, sell, or hold a stock based on technical indicators


```python
current_stock_price=float(input("Enter current stock price: "))
fifty_day_moving_average=float(input("Enter 50-day moving average: "))
RSI_value=float(input("Enter RSI value (0-100): "))
current_portfolio_cash=float(input("Enter current portfolio cash: "))
if current_stock_price>fifty_day_moving_average and RSI_value<30 and current_portfolio_cash>current_stock_price:
    print("Action: BUY")
elif current_stock_price<fifty_day_moving_average or RSI_value>70:
    print("Action: SELL")
else:
    print("Action: HOLD")    
```

#### Automated Cyber-Threat Firewall   ####
Act as a network security firewall evaluating whether to allow, flag, or block incoming web traffic packets.

```python
# Setup local home country configuration
HOME_COUNTRY = "India"

# 1. Gather Packet Inputs
source_country = input("Enter Source IP Country: ").strip().title()
connection_attempts = int(input("Enter Connection Attempts per minute: "))
packet_size = float(input("Enter Packet Size in MB: "))
is_known_malicious_ip = input("Is Known Malicious IP? (True/False): ").strip().lower().capitalize() == "True"

print("-" * 40)

# 2. Firewall Evaluation Logic
# Rule 1 & 2: Immediate Blocks
if is_known_malicious_ip:
    print("TRAFFIC BLOCKED (Reason: Malicious IP)")
elif connection_attempts > 200:
    print("TRAFFIC BLOCKED (Reason: DDoS Mitigation)")

# Rule 3 & 4: Flags for Review
elif source_country != HOME_COUNTRY and connection_attempts > 50:
    print("TRAFFIC FLAGGED (Reason: Foreign high-frequency traffic)")
elif packet_size > 500.0:
    print("TRAFFIC FLAGGED (Reason: Excessive packet size)")

# Rule 5: Catch-all safe traffic
else:
    print("TRAFFIC ALLOWED")
```
