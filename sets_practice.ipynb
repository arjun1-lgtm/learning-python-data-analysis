Question 6: The Copycat
What happens to set_x in the following code when we mutate set_y?

set_x = {1, 2, 3}
set_y = set_x
set_y.add(4)
print(set_x)

```python
#  It actually prints {1, 2, 3, 4}.

'''Here is why: doing set_y = set_x does not create a new copy of the set. 
Instead, it just creates a new nickname pointing to the exact same set in your computer's memory.'''
```
