```python
"""Write a for loop that iterates through a list of names: ["Alice", "Bob", "Charlie"].
 If the name is "Bob", use the pass statement (maybe you plan to add special logic for Bob later). For any other name, print "Hello, [Name]"."""

names=["Alice", "Bob", "Charlie"]
for name in names:
    if name=="Bob":
        pass
    else:
        print(name)
```

```python
"""Create an infinite while True loop. 
Inside, ask the user for input. If the user types "stop", break out of the loop. Otherwise, keep asking them."""


while True:
    user=input("Enter ('stop'or 'anything other than stop') to control the loop: ").lower()
    if user=="stop":
        print("Loop stopped.")
        break
    print(user)
```

```python
'''Write a program that calculates the sum of all numbers from 1 to 50 using a loop,
 but skips any number that is a multiple of 5. Print the final total at the end.'''


total=0
for i in range(1,51):
    if i%5==0:
        continue
    total+=i
print(total)
```

#### 1. The Password Validator #### 
Write a program that keeps asking a user to create a password using a while True loop. The password must meet all of these conditions:

It must be at least 8 characters long.

It must contain at least one number.

If it meets the conditions, print "Password secure!" and break. Otherwise, print what requirement they missed and let them try again.

Hint: You can use .isdigit() to check if a character in a string is a number. 

```python
while True:
    passw=input("create a password(at least 8 characters long and must contain at least one digit):")
    if len(passw)>=8 and any(char.isdigit() for char in passw):
        print("Password secure!.")
        break
    elif len(passw)<8 and not any(char.isdigit() for char in passw):
        print("Password must be at least 8 characters long and contain at least one digit.")
    elif len(passw)<8:
        print("Password must be at least 8 characters long.")
    else:
        print("Password must contain at least one digit.")    
```

#### 2.FizzBuzz (for + multiple conditions) ####
Loop through numbers from 1 to 30.

If a number is divisible by 3, print "Fizz".

If a number is divisible by 5, print "Buzz".

If it is divisible by both 3 and 5, print "FizzBuzz".

Otherwise, just print the number itself.

```python
for i in range(1,31):
    if i%3==0 and i%5==0:
        print("FizzBuzz")
    elif i%3==0:
        print("Fizz")
    elif i%5==0:
        print("Buzz")
    else:
        print(i)
        
```

####  3. The Duplicate Destroyer (List filtering) ####
Given a list with duplicate items: [1, 2, 2, 3, 4, 4, 4, 5, 6, 1].

Create a brand new empty list called unique_list. Use a loop to iterate through the original list, and populate unique_list so that it contains each number exactly once.

Hint: Use the in or not in keywords to check if a number has already been added to your new list before you append it!

```python
a=[1, 2, 2, 3, 4, 4, 4, 5, 6, 1]
b=[]
for i in a:
    if i not in b:
        b.append(i)
print(b)
```

```python
#easy method
b=list(set(a))
print(b)
```

#### 4. Find the Peak (Tracking maximums) ####
Without using Python's built-in max() function, write a loop to find the largest number in this list:
nums = [14, 27, 8, 42, 35, 21]
Print the largest number at the very end.

```python
nums = [14, 27, 8, 42, 35, 21]
j=0
for i in nums:
    if i>j:
        j=i
print(j)

'''this code is correct only if your list contains only positive numbers. If there are negative numbers, 
you should initialize j with a very small number or the first element of the list.
so if you want to find the largest number in a list that may contain negative numbers,
 you can initialize j with the first element of the list:'''

nums = [14, 27, 8, 42, 35, 21, -5, -10]
j=nums[0]                                          # this will work as it initializes j with the first element of the list and the first largest number, which is 14 in this case.
for i in nums:
    if i>j:
        j=i
print(j)
```

#### 5.Nested Star Pattern (Nested loops) ####
Write a nested loop (a loop inside another loop) to print the following shape of stars (*):

```python
for i in range(1,6):
    for j in range(i):
        print("*",end="")
    print()
```
